```python
def process(x):
    x=str(x).replace(',','.')
    if str(x)[-1]=='%':
        return float(x[:-1])/100
    else:
        try:
            return float(x)
        except Exception as e:
            return 0.0
fullTable=fullTable.applymap(process)
fullTable=fullTable.fillna(0)

table27=fullTable[[x for x in fullTable.columns if '27' in x]]
table26=fullTable[[x for x in fullTable.columns if '26' in x]]

table26.to_csv('./data/26.csv')
table27.to_csv('./data/27.csv')
#table26;
#table27;
```
